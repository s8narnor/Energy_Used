##**Feature Transformation**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_labels, Enguse_train, test_size = 0.2)
```

```python
# Dimensionality Reduction of Train Data

from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA

normalized_data = X_train.copy(deep=True)

# Scale the train data using the natural logarithm
normalized_data =np.log(normalized_data)

n_components =4

# Apply PCA by fitting the normalized data with 4 dimentions
pca = PCA(n_components = n_components)
reduced_data = pca.fit_transform(normalized_data)
reduced_data = pd.DataFrame(reduced_data, columns = ['Dimension' + str(i+1) for i in range(n_components)])


variance = pd.DataFrame(pca.explained_variance_ratio_)
print(np.cumsum(pca.explained_variance_ratio_))


# Dimensionality Reduction of Test Data
normalized_data_test = X_test.copy(deep=True)


# Scale the test data using the natural logarithm
normalized_data_test =np.log(normalized_data_test)

# Transform normalized_data_test using pca fit
reduced_data_test = pca.transform(normalized_data_test)
reduced_data_test = pd.DataFrame(reduced_data_test, columns = ['Dimension' + str(i+1) for i in range(n_components)])
```

##**Regression Algorithms**

```python
# RUL regression model selection

from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge, Lasso, ElasticNet, BayesianRidge, HuberRegressor
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor, AdaBoostRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor

from sklearn.metrics import make_scorer, mean_squared_error
from sklearn.model_selection import KFold, cross_val_score

def analyse_regression_models(train, labels):
    results = {}
    
    def test_model(clf):
        cv = KFold(n_splits = 2, shuffle = True, random_state = 20)
        mse = make_scorer(mean_squared_error)
        mse_val_score = cross_val_score(clf, train, labels, cv=cv, scoring = mse)
        scores = [mse_val_score.mean()]
        return scores
    
    clf = LinearRegression()
    results['LinearRegression'] = test_model(clf)

    clf = Ridge()
    results['Ridge'] = test_model(clf)

    clf = Lasso()
    results['Lasso'] = test_model(clf)

    clf = ElasticNet()
    results['ElasticNet'] = test_model(clf)

    clf = BayesianRidge()
    results['BayesianRidge'] = test_model(clf)
    
    clf = HuberRegressor()
    results['HuberRegressor'] = test_model(clf)
    
    clf = MLPRegressor()
    results['NeuralNetwork'] = test_model(clf)
    
    clf = BaggingRegressor()
    results['BaggingRegressor'] = test_model(clf)
    
    clf = RandomForestRegressor()
    results['RandomForestRegressor'] = test_model(clf)
    
    clf = AdaBoostRegressor()
    results['AdaBoostRegressor'] = test_model(clf)
    
    clf = SVR()
    results['SVM'] = test_model(clf)
    
    results = pd.DataFrame.from_dict(results, orient = 'index')
    results.columns = ["Mean Squared Error"]
    results = results.sort_values(by = ["Mean Squared Error"], ascending = True)
    results.plot(kind = 'bar', title = "Model Scores")
    axes = plt.gca()
    axes.set_ylim([0,2000])
    return results

analyse_regression_models(reduced_data, y_train)
    
```

###**Random Forest**

```python
# Train and Test the selected model for RUL (Regression)
# Predict RUL for turbofan engine using Regression

from sklearn.metrics import mean_squared_error, make_scorer, mean_absolute_error
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.ensemble import RandomForestRegressor

cv = KFold(n_splits =2, shuffle = True, random_state =20)

# Parameter list for tunning
parameters = {'bootstrap': [True], 'criterion': ['mse'],
              'max_depth': [4], 
              'max_features': ['auto'], 
              'max_leaf_nodes':[None], 
              'min_impurity_decrease':[0.0],
              'min_impurity_split':[None],
              'min_samples_leaf': [2],
              'min_samples_split': [2],
              'min_weight_fraction_leaf':[0.0],
              'n_estimators': [130],
              'n_jobs': [-1],
              'oob_score': [False],
              'random_state':[20],
              'verbose': [0],
              'warm_start': [False]}

# Initialise RandomForestRegressor
clf = RandomForestRegressor(n_estimators=100, criterion= 'mse', max_depth=None, min_samples_split=2)
mse = make_scorer(mean_squared_error)

# Perform grid search using mse as the scoring method
grid_obj = GridSearchCV(clf, parameters, cv=cv, scoring =mse)

#Fit the grid search to the training data and find the optimal parameters
grid_fit = grid_obj.fit(reduced_data, y_train)


# Get the best estimator
best_clf = grid_fit.best_estimator_
print(best_clf)

best_clf.fit(reduced_data, y_train)

Eng_predictions = best_clf.predict(reduced_data_test)

print("Mean Absolute Error: " + str(mean_absolute_error(Eng_predictions, y_test)))
print("Mean Squared Error: " + str(mean_squared_error(Eng_predictions, y_test)))


# Plot the predicted RUL values against the actual RUL values
plt.scatter(y_test, Eng_predictions)

# Line of perfect fit
straight_line = np.arange(0,250)
plt.plot(straight_line, straight_line)
plt.title("Fitted Values")
```

####**Save the model**

```python
import pickle
  
# Save the trained model as a pickle string.
rf_model = pickle.dumps(best_clf)
  
# Load the pickled model
rf_from_pickle = pickle.loads(rf_model)
  
```
